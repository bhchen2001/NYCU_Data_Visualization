```python
import pandas as pd
```

```python
# read the data
df = pd.read_csv('../dataset/time_dependent_data_A1.csv')
```

```python
# split the time into year, month, day, hour, minute, second
df['year'] = df['Date_ID'].apply(lambda x: str(x)[:4])
df['month'] = df['Date_ID'].apply(lambda x: str(x)[4:6])
df['day'] = df['Date_ID'].apply(lambda x: str(x)[6:8])
df['hour'] = df['Date_ID'].apply(lambda x: str(x)[8:10])
df['minute'] = df['Date_ID'].apply(lambda x: str(x)[10:12])
df['second'] = df['Date_ID'].apply(lambda x: str(x)[12:14])
```

```python
# add day of week
df['day_of_week'] = pd.to_datetime(df['Date_ID'], format='%Y%m%d%H%M%S').dt.dayofweek
```

```python
# count num of rows for each day of week
df.groupby('day_of_week').count()['Date_ID']
```

```python
# split time from 00:00 to 23:59 into 6 slots
df['time_slot'] = df['hour'].apply(lambda x: int(x) // 4)
```

```python
df.groupby('time_slot').count()['hour']
```

```python
# for each day of week, count the number of rows for each time slot
df.groupby(['day_of_week', 'time_slot']).count()['Date_ID']
```

```python
# write the count result to json
accident_list = []
for i in range(7):
    for j in range(6):
        tmp_dict = {
            'day_of_week': i,
            'time_slot': j,
            'count': df.groupby(['day_of_week', 'time_slot']).count()['Date_ID'][i][j]
        }
        accident_list.append(tmp_dict)
```
